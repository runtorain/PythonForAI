```python
# prompt: plot normal curve of age

import numpy as np
import scipy.stats as stats

# Extract the 'Age' column and remove missing values
age_data = df['Age'].dropna()

# Fit a normal distribution to the age data
mu, std = stats.norm.fit(age_data)

# Generate x-values for the plot
x = np.linspace(age_data.min(), age_data.max(), 100)

# Calculate the corresponding y-values (probability density function)
y = stats.norm.pdf(x, mu, std)

# Plot the normal curve
plt.plot(x, y)
plt.xlabel('Age')
plt.ylabel('Probability Density')
plt.title('Normal Curve of Age')
plt.show()
```

```python
# prompt: plot normal curve of Fare

import numpy as np
import scipy.stats as stats

# Extract the 'Fare' column and remove missing values
Fare_data = df['Fare'].dropna()

# Fit a normal distribution to the Fare data
mu, std = stats.norm.fit(Fare_data)

# Generate x-values for the plot
x = np.linspace(Fare_data.min(), Fare_data.max(), 100)

# Calculate the corresponding y-values (probability density function)
y = stats.norm.pdf(x, mu, std)

# Plot the normal curve
plt.plot(x, y)
plt.xlabel('Fare')
plt.ylabel('Probability Density')
plt.title('Normal Curve of Fare')
plt.show()
```
